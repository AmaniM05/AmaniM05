```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.isitdownrightnow.com/canada.ca.html'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
data = soup.find('table')
```

```python
data
```

```python
data = []
for row in soup.find_all("tr")[1:]:
    cells = row.find_all("td")
    date_1 = cells[0].get_text(strip=True).replace('\xa0', '')
    time_1 = cells[1].get_text(strip=True)
    date_2 = cells[4].get_text(strip=True).replace('\xa0', '')
    time_2 = cells[5].get_text(strip=True)
    data.append((date_1, time_1))
    data.append((date_2, time_2))
```

```python
for date, time in data:
    print(f"Date: {date}, Time: {time}")
```
